### pickle

2.1 На основе файла `contributors_sample.json` создайте словарь следующего вида:
```
{
    должность: [список username людей, занимавших эту должность]
}
```

````python
with open('contributors_sample.json', mode='r') as f:  # Открытие файла 'contributors_sample.json' для чтения ('r') и создание файла объекта 'f'.
    contributors_sample = json.load(f)  # Загрузка данных из файла JSON в переменную 'contributors_sample'.

jobs = []  # Создание пустого списка 'jobs' для хранения всех должностей.

for i in range(len(contributors_sample)):  # Цикл for для каждого элемента в переменной 'contributors_sample'.
    for j in range(len(contributors_sample[i]['jobs'])):  # Вложенный цикл for для каждого элемента в списке значений ключа 'jobs' внутри элемента 'i'.
        jobs.append(contributors_sample[i]['jobs'][j])  # Добавление каждой должности в список 'jobs'.

pos = []  # Создание пустого списка 'pos' для хранения словарей должностей и соответствующих им пользователей.

for i in range(len(jobs)):  # Цикл for для каждой должности в списке 'jobs'.
    pos.append({jobs[i]: [el['username'] for el in contributors_sample if jobs[i] in el['jobs']]})  # Создание словаря, где ключ - должность, а значение - список пользователей с этой должностью.

pos  # Возврат списка 'pos', содержащего словари должностей и соответствующих им пользователей.
````

3.1 По данным файла `steps_sample.xml` сформируйте словарь с шагами по каждому рецепту вида `{id_рецепта: ["шаг1", "шаг2"]}`. Сохраните этот словарь в файл `steps_sample.json`

````python
with open('/content/steps_sample.xml') as f:  # Открытие файла 'steps_sample.xml' для чтения и создание файла объекта 'f'.
    ab = bs(f, 'xml')  # Парсинг XML-данных с помощью BeautifulSoup и сохранение результата в переменную 'ab'.

ids = [id.text for id in [ab.find_all('id')][0]]  # Получение текстового содержимого всех элементов 'id' и сохранение в списке 'ids'.

recipes = [ab.find_all('steps')][0]  # Получение всех элементов 'steps' и сохранение в списке 'recipes'.

res = {}  # Создание пустого словаря 'res' для хранения результатов.

steps = []  # Создание пустого списка 'steps' для хранения шагов рецептов.

for recipe in recipes:  # Цикл for для каждого элемента 'recipe' в списке 'recipes'.
    steps.append([i.text for i in recipe.find_all('step')])  # Для каждого элемента 'recipe' получаем текстовое содержимое всех элементов 'step' и добавляем в список 'steps'.

res = {ids[i]: steps[i] for i in range(len(ids))}  # Создание словаря 'res' с id в качестве ключей и списками шагов в качестве значений.

with open('/content/steps_sample.json', 'w') as file:  # Открытие файла 'steps_sample.json' для записи и создание файла объекта 'file'.
    json.dump(res, file, indent=4)  # Запись словаря 'res' в JSON-формат с отступами в файл.
````

3.2 По данным файла `steps_sample.xml` сформируйте словарь следующего вида: `кол-во_шагов_в_рецепте: [список_id_рецептов]`

````python
n_steps = {ids[i]: len(steps[i]) for i in range(len(ids))}  # Создание словаря 'n_steps' с id в качестве ключей и количеством шагов в качестве значений.

dict_steps = {}  # Создание пустого словаря 'dict_steps' для хранения id рецептов в зависимости от количества шагов.

for i, j in n_steps.items():  # Цикл for для каждой пары ключ-значение в словаре 'n_steps'.
    if j in dict_steps.keys():  # Проверка, есть ли значение j в качестве ключа в словаре 'dict_steps'.
        dict_steps[j].append(i)  # Если значение j уже есть в словаре, добавляем текущий id 'i' к списку значений.
    else:
        dict_steps[j] = [i]  # Если значение j еще не существует как ключ, создаем новую запись с ключом j и значением, равным списку, содержащему только текущий id 'i'.

print(dict_steps)  # Вывод словаря 'dict_steps', содержащего количество шагов в качестве ключей и список соответствующих им id рецептов.
````

3.3 Получите список рецептов, в этапах выполнения которых есть информация о времени (часы или минуты). Для отбора подходящих рецептов обратите внимание на атрибуты соответствующих тэгов.

````python
res = {}  # Создание пустого словаря 'res' для хранения результатов.

ids = [id.text for id in [ab.find_all('id')][0]]  # Получение текстового содержимого всех элементов 'id' и сохранение в списке 'ids'.

recipes = [len(recipe) for recipe in [ab.find_all('steps')][0]]  # Получение количества элементов в каждом элементе 'steps' и сохранение в списке 'recipes'.

for num, recipe in enumerate(recipes):  # Цикл for с использованием функции enumerate для перебора индексов и значений в списке 'recipes'.
    if res.get(recipe) == None:  # Проверка, существует ли ключ с количеством шагов 'recipe' в словаре 'res'.
        res[recipe] = [ids[num]]  # Если ключа нет, создаем новую запись в словаре с ключом 'recipe' и значением, равным списку, содержащему текущий id 'ids[num]'.
    else:
        res[recipe].append(ids[num])  # Если ключ уже существует, добавляем текущий id 'ids[num]' к списку значений.

res  # Возврат словаря 'res' с количеством шагов в качестве ключей и списками соответствующих им id рецептов.
````

3.4 Загрузите данные из файла `recipes_sample.csv` (__ЛР2__) в таблицу `recipes`. Для строк, которые содержат пропуски в столбце `n_steps`, заполните этот столбец на основе файла  `steps_sample.xml`. Строки, в которых столбец `n_steps` заполнен, оставьте без изменений.

````python
recipes = pd.read_csv('/content/recipes_sample.csv', delimiter=',', parse_dates=['submitted'])  # Загрузка данных из CSV-файла в DataFrame 'recipes', указывая разделитель ',' и парсинг столбца 'submitted' как дату.

for index, row in recipes.iterrows():  # Цикл for для итерации по каждой строке DataFrame 'recipes'.
    if pd.isnull(row['n_steps']):  # Проверка, является ли значение столбца 'n_steps' в текущей строке пустым.
        recipe_id = str(row['id'])  # Получение id рецепта из столбца 'id' текущей строки и преобразование в строковый тип.
        if recipe_id in n_steps:  # Проверка, существует ли количество шагов для данного id рецепта в словаре 'n_steps'.
            recipes.at[index, 'n_steps'] = n_steps[recipe_id]  # Если количество шагов найдено, устанавливаем значение столбца 'n_steps' в DataFrame 'recipes' равным этому значению.

recipes  # Возврат обновленного DataFrame 'recipes' с добавленными значениями количества шагов.
````
